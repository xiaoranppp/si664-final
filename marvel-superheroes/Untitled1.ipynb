```python
power=[]
with open('superheroes_power_matrix.csv') as csv_file4:
    csv_reader4 = csv.reader(csv_file4, delimiter=',')
    for row in csv_reader4:
        power.append(row)
```

```python
powertable=[]
for i in range(1,len(power[0])):
    powertable.append([i,power[0][i]])
    
```

```python
name=['Id','Power']
test=pd.DataFrame(columns=name,data=powertable)#数据有三列，列名分别为one,two,three
test.to_csv('powertable.csv',encoding='gbk')
```

```python
ctp=[]
for p in havepower:
    for k in range(1,len(p)):
        if p[k]=='TRUE':
            ctp.append([p[0],powertable[k-1][1]])
            
```
