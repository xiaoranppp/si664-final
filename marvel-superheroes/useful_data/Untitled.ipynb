```python
import pandas as pd
character=pd.read_csv('character_total2.csv')
```

```python
def trim_columns(data_frame):
	"""
	:param data_frame:
	:return: trimmed data frame
	"""
	trim = lambda x: x.strip() if type(x) is str else x
	return data_frame.applymap(trim)
trim_columns(character)
```

```python
character.to_csv('character_total3.csv',encoding='utf-8')
```
